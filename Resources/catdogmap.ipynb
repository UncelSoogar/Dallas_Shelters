```python
#catdog[catdog["Census Tract"]!="W"]
#catdog[catdog["Census Tract"].apply(lambda x: x.isnumeric())]
#cleaning letters out
catdog = catdog[pd.to_numeric(catdog['Census Tract'], errors='coerce').notnull()]
catdog
```

```python
catdog['Census Tract'] = pd.to_numeric(catdog['Census Tract'])
```

```python
catdog["Census Tract"].value_counts()
```

```python
catdog["Census Tract"] = catdog["Census Tract"]/100
```
